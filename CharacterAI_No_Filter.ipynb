# ****Cara mendownload karakter**** (.json)

1. Download Discord
2. [Klik Ini Untuk Join Discord Server](discord.gg/pygmalionai) Atau jika tidak bisa di klik, Join dengan link https://discord.gg/pygmalionai
3. Verify disana
4. Pergi ke Channel Character Sharing dan download .json nya
5. Upload Karakter kalian

#****Cara Chat Bareng Karakter****

1. Klik link yang sudah di generate oleh Google Colab (Yang Public URL)

2. Download atau Buat json Karakter yang kamu mau (Untuk bikin bisa [Klik Disini](https://oobabooga.github.io/character-creator.html))

3. Setelah itu, pergi ke website dengan link yang kalian buka tadi dan pergi ke ***Menu Karakter (Samping Text Generation)***

4. Di menu karakter (Character), Scroll kebawah sampai ketemu tombol Chat History Dan Upload Character dan ***Klik Upload Character***

5. Di nomor pertama (1. Select the JSON file), ***Upload file .json Karakter kalian yang sudah kalian buat/download***

6. (Opsional) Upload Foto Karakter kalian di nomor kedua (2. Select your character’s profile picture) 

7. Klik ***Submit***

8. Tepat diatas menu Upload Character, ***Klik ikon refresh (🔁)***

9. Tepat sebelah kiri ikon refresh, ***Klik Character dan pilih nama karakter kalian***

10. Kalian scroll ke atas dan pergi ke ***Text Generation***

11. Dibagian `input` itu adalah Text yang akan kalian kirim ke Karakter kalian. Untuk membuat karakter kalian merespon, ***Klik Generate untuk mengirimkan Text kalian***

12. Selamat Chat!
